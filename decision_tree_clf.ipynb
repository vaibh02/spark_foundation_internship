## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import confusion_matrix
```

## Loading the Data 

```python
a=load_iris()
```

```python
X=a.data
y=a.target
```

```python
a.target_names
```

```python
a.target
```

```python
a.feature_names
```

```python
X[:5]
```

```python
df=pd.DataFrame(a.data,columns=a.feature_names)
df['species']=a.target_names[a.target]
```

```python
df
```

```python
plt.hist('species',data=df)
plt.show()
```

## Data Spliting into train/test sets

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.27,random_state=1)
```

```python
X_train[:10]
```

## Model Definition

```python
tree=DecisionTreeClassifier(random_state=1)
```

```python
tree.fit(X_train,y_train)
```

## Evaluation of the Model

```python
tree.score(X_test,y_test)
```

## Confusion Matrix

```python
cm=confusion_matrix(y_test,tree.predict(X_test))
```

```python
sns.heatmap(cm,annot=True)
```

# Decision Tree in Visual Form 

```python
plt.figure(figsize=(10,10))
plot_tree(tree,feature_names=a.feature_names,filled=True,proportion=True)
plt.show()
```
